```python
df_numerica = df_filtrado.select_dtypes(include=np.number)

correlation_matrix = df_numerica.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlacin de Variables Numricas')
plt.show()
```

```python
corr_par = correlation_matrix.unstack()
strongest_pairs_excluding_self = corr_par[(corr_par.index.get_level_values(0) != corr_par.index.get_level_values(1))]
sorted_corr_par = strongest_pairs_excluding_self.sort_values(kind="quicksort")

print("\nPar de variables con mayor correlacin positiva:")
print(sorted_corr_par.tail(1).round(2))
print("\nPar de variables con mayor correlacin negativa:")
print(sorted_corr_par.head(1).round(2))
```

## <font color='green'>Conclusiones:</font></h1></center>

**1. Mayor Correlacin Positiva (0.98)**

Variables: **Tasa de Fertilidad y Tasa de Natalidad**

**Interpretacin:** Existe una correlacin positiva casi perfecta entre estas dos variables.

Esto significa que en los pases donde la tasa de fertilidad (nmero promedio de hijos por mujer) es alta, la tasa de natalidad (nacimientos por cada 1,000 habitantes) tambin tiende a ser alta, y viceversa.

Por qu ocurre esto?

Relacin demogrfica directa:

La tasa de fertilidad mide la cantidad de hijos por mujer en edad reproductiva.

La tasa de natalidad mide cuntos nacimientos ocurren en una poblacin.

Ambas estn intrnsecamente ligadas: a ms hijos por mujer (fertilidad alta), ms nacimientos habr (natalidad alta).

**2. Mayor Correlacin Negativa (-0.92)**

Variables: **Mortalidad Infantil y Esperanza de Vida**

**Interpretacin:** Hay una fuerte correlacin negativa entre estas variables.

Esto indica que, en general, los pases con alta mortalidad infantil tienden a tener una esperanza de vida ms baja, y viceversa.

Por qu ocurre esto?

Indicadores de desarrollo: La mortalidad infantil refleja la calidad del sistema de salud, acceso a agua potable, nutricin y condiciones sanitarias.

La esperanza de vida tambin depende de estos factores, pero adems de otros como enfermedades crnicas, violencia, etc.

Causalidad indirecta: Una alta mortalidad infantil reduce el promedio de esperanza de vida (porque muchos nios mueren jvenes).

Pero incluso si se excluyeran esas muertes, los pases con alta mortalidad infantil suelen tener peor salud en general, lo que tambin reduce la esperanza de vida en adultos.

Sin embargo, que la relacin no sea perfecta (-1.0), sugiere que hay otros factores influyendo (ej: un pas podra mejorar la esperanza de vida con tratamientos para adultos, pero an tener alta mortalidad infantil).

```python
df_gdp_pop = df[['Country', 'Population', 'GDP']].copy()

df_gdp_pop['Population'] = df_gdp_pop['Population'].astype(str).str.replace(',', '', regex=False)
df_gdp_pop['Population'] = pd.to_numeric(df_gdp_pop['Population'], errors='coerce')

df_gdp_pop['GDP'] = df_gdp_pop['GDP'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df_gdp_pop['GDP'] = pd.to_numeric(df_gdp_pop['GDP'], errors='coerce')

df_gdp_pop_cleaned = df_gdp_pop.dropna(subset=['Population', 'GDP']).copy()

print("\nDataFrame con Poblacin y PIB limpios:")
print(df_gdp_pop_cleaned.info())
print(df_gdp_pop_cleaned.head())
```

Seleccin de Grfico: **Grfico de dispersin**

Queremos visualizar la relacin entre dos variables cuantitativas continuas: Poblacin y PIB.

Un grfico de dispersin (scatterplot) es el ms adecuado para este propsito, ya que nos permite ver si existe una correlacin entre las dos variables y cmo se distribuyen los puntos de datos.

Dado que tanto la Poblacin como el PIB pueden variar en varios rdenes de magnitud entre pases, es til usar una escala logartmica en ambos ejes para poder visualizar mejor la relacin para pases con valores tanto bajos como altos, evitando que los pases grandes dominen el grfico y oculten los detalles de los ms pequeos.

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_gdp_pop_cleaned, x='Population', y='GDP', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Relacin entre Poblacin Total y Producto Interno Bruto (PIB) por Pas', fontsize=16)
plt.xlabel('Poblacin Total', fontsize=12)
plt.ylabel('Producto Interno Bruto (PIB)', fontsize=12)
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

## <font color='green'>Conclusiones:</font></h1></center>

El grfico de dispersin con ejes logartmicos muestra una clara correlacin positiva entre la Poblacin y el PIB: los pases con mayor poblacin tienden a tener un PIB ms alto. Sin embargo, la dispersin de los puntos tambin indica que la relacin no es perfectamente lineal y que existen variaciones significativas en el PIB para pases con poblaciones similares, lo que sugiere la influencia de otros factores econmicos y estructurales ms all del simple tamao de la poblacin. Los pases se agrupan principalmente en la parte inferior izquierda del grfico, con algunos valores atpicos hacia la parte superior derecha (pases con poblaciones y PIB muy elevados).

Resumen:

- Se us escala logartmica debido a la gran variacin en los valores.

- Se eliminaron outliers extremos que podran distorsionar la visualizacin.

- Se observa una relacin positiva pero no perfecta entre poblacin y PIB.

Analizaremos la relacin entre **Tasa de Fertilidad** y **Tasa de Mortalidad Infantil:**

```python
df_fert_mort = df_filtrado[['Country', 'Fertility Rate', 'Infant mortality']].dropna()

df_fert_mort['Fertility Rate'] = pd.to_numeric(df_fert_mort['Fertility Rate'], errors='coerce')
df_fert_mort['Infant mortality'] = pd.to_numeric(df_fert_mort['Infant mortality'], errors='coerce')
df_fert_mort = df_fert_mort.dropna()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_fert_mort, x='Fertility Rate', y='Infant mortality')
plt.title('Tasa de Fertilidad vs Mortalidad Infantil')
plt.xlabel('Tasa de Fertilidad (hijos por mujer)')
plt.ylabel('Mortalidad Infantil (por 1000 nacimientos)')
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(data=df_fert_mort, x='Fertility Rate', y='Infant mortality')
plt.title('Relacin entre Tasa de Fertilidad y Mortalidad Infantil')
plt.xlabel('Tasa de Fertilidad')
plt.ylabel('Mortalidad Infantil')
plt.show()

correlation = df_fert_mort['Fertility Rate'].corr(df_fert_mort['Infant mortality'])
print(f"Correlacin entre Tasa de Fertilidad y Mortalidad Infantil: {correlation:.2f}")
```

## <font color='green'>Conclusiones:</font></h1></center>

- Existe una fuerte correlacin positiva entre la tasa de fertilidad y la mortalidad infantil.

- Pases con mayor tasa de fertilidad tienden a tener mayor mortalidad infantil, lo que podra indicar diferencias en desarrollo socioeconmico y acceso a salud.
