```python
df_numerica = df_filtrado.select_dtypes(include=np.number)

correlation_matrix = df_numerica.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```

```python
corr_par = correlation_matrix.unstack()
strongest_pairs_excluding_self = corr_par[(corr_par.index.get_level_values(0) != corr_par.index.get_level_values(1))]
sorted_corr_par = strongest_pairs_excluding_self.sort_values(kind="quicksort")

print("\nPar de variables con mayor correlación positiva:")
print(sorted_corr_par.tail(1).round(2))
print("\nPar de variables con mayor correlación negativa:")
print(sorted_corr_par.head(1).round(2))
```

## <font color='green'>Conclusiones:</font></h1></center>

**1. Mayor Correlación Positiva (0.98)**

Variables: **Tasa de Fertilidad y Tasa de Natalidad**

**Interpretación:** Existe una correlación positiva casi perfecta entre estas dos variables.

Esto significa que en los países donde la tasa de fertilidad (número promedio de hijos por mujer) es alta, la tasa de natalidad (nacimientos por cada 1,000 habitantes) también tiende a ser alta, y viceversa.

¿Por qué ocurre esto?

Relación demográfica directa:

La tasa de fertilidad mide la cantidad de hijos por mujer en edad reproductiva.

La tasa de natalidad mide cuántos nacimientos ocurren en una población.

Ambas están intrínsecamente ligadas: a más hijos por mujer (fertilidad alta), más nacimientos habrá (natalidad alta).

**2. Mayor Correlación Negativa (-0.92)**

Variables: **Mortalidad Infantil y Esperanza de Vida**

**Interpretación:** Hay una fuerte correlación negativa entre estas variables.

Esto indica que, en general, los países con alta mortalidad infantil tienden a tener una esperanza de vida más baja, y viceversa.

¿Por qué ocurre esto?

Indicadores de desarrollo: La mortalidad infantil refleja la calidad del sistema de salud, acceso a agua potable, nutrición y condiciones sanitarias.

La esperanza de vida también depende de estos factores, pero además de otros como enfermedades crónicas, violencia, etc.

Causalidad indirecta: Una alta mortalidad infantil reduce el promedio de esperanza de vida (porque muchos niños mueren jóvenes).

Pero incluso si se excluyeran esas muertes, los países con alta mortalidad infantil suelen tener peor salud en general, lo que también reduce la esperanza de vida en adultos.

Sin embargo, que la relación no sea perfecta (-1.0), sugiere que hay otros factores influyendo (ej: un país podría mejorar la esperanza de vida con tratamientos para adultos, pero aún tener alta mortalidad infantil).

```python
par1_var1 = 'Fertility Rate'
par1_var2 = 'Birth Rate'

df_par1 = df_filtrado[[par1_var1, par1_var2]].dropna().copy()

print(f"\nDimensiones del DataFrame para el par 1 después de eliminar faltantes: {df_par1.shape}")
print(f"Cantidad de faltantes en par 1 después de eliminar faltantes:\n{df_par1.isnull().sum()}")
```

```python
Q1_par1 = df_par1.quantile(0.25)
Q3_par1 = df_par1.quantile(0.75)
IQR_par1 = Q3_par1 - Q1_par1

lower_bound_par1 = Q1_par1 - 1.5 * IQR_par1
upper_bound_par1 = Q3_par1 + 1.5 * IQR_par1

df_par1_cleaned = df_par1[
    (df_par1[par1_var1] >= lower_bound_par1[par1_var1]) & (df_par1[par1_var1] <= upper_bound_par1[par1_var1]) &
    (df_par1[par1_var2] >= lower_bound_par1[par1_var2]) & (df_par1[par1_var2] <= upper_bound_par1[par1_var2])].copy()

print(f"\nDimensiones del DataFrame para el par 1 después de eliminar atípicos: {df_par1_cleaned.shape}")
print(f"Cantidad de filas eliminadas (posibles atípicos) para el par 1: {df_par1.shape[0] - df_par1_cleaned.shape[0]}")
```

```python
par2_var1 = 'Infant mortality'
par2_var2 = 'Life expectancy'

df_par2 = df_filtrado[[par2_var1, par2_var2]].dropna().copy()
print(f"\nDimensiones del DataFrame para el par 2 después de eliminar faltantes: {df_par2.shape}")
print(f"Cantidad de faltantes en par 2 después de eliminar faltantes:\n{df_par2.isnull().sum()}")
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_par1_cleaned[par1_var1], y=df_par1_cleaned[par1_var2])
plt.title(f'Relación entre {par1_var1} y {par1_var2} (sin faltantes ni atípicos)')
plt.xlabel(par1_var1)
plt.ylabel(par1_var2)
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_par2_cleaned[par2_var1], y=df_par2_cleaned[par2_var2])
plt.title(f'Relación entre {par2_var1} y {par2_var2} (sin faltantes ni atípicos)')
plt.xlabel(par2_var1)
plt.ylabel(par2_var2)
plt.show()

correlation_par1_cleaned = df_par1_cleaned.corr().iloc[0, 1]
correlation_par2_cleaned = df_par2_cleaned.corr().iloc[0, 1]

print(f"\nCorrelación entre {par1_var1} y {par1_var2} (después de limpieza): {correlation_par1_cleaned:.2f}")
print(f"Correlación entre {par2_var1} y {par2_var2} (después de limpieza): {correlation_par2_cleaned:.2f}")
```

```python
df_gdp_pop = df[['Country', 'Population', 'GDP']].copy()

df_gdp_pop['Population'] = df_gdp_pop['Population'].astype(str).str.replace(',', '', regex=False)
df_gdp_pop['Population'] = pd.to_numeric(df_gdp_pop['Population'], errors='coerce')

df_gdp_pop['GDP'] = df_gdp_pop['GDP'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df_gdp_pop['GDP'] = pd.to_numeric(df_gdp_pop['GDP'], errors='coerce')

df_gdp_pop_cleaned = df_gdp_pop.dropna(subset=['Population', 'GDP']).copy()

print("\nDataFrame con Población y PIB limpios:")
print(df_gdp_pop_cleaned.info())
print(df_gdp_pop_cleaned.head())
```

Selección de Gráfico: **Gráfico de dispersión**

Queremos visualizar la relación entre dos variables cuantitativas continuas: Población y PIB.

Un gráfico de dispersión (scatterplot) es el más adecuado para este propósito, ya que nos permite ver si existe una correlación entre las dos variables y cómo se distribuyen los puntos de datos.

Dado que tanto la Población como el PIB pueden variar en varios órdenes de magnitud entre países, es útil usar una escala logarítmica en ambos ejes para poder visualizar mejor la relación para países con valores tanto bajos como altos, evitando que los países grandes dominen el gráfico y oculten los detalles de los más pequeños.

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_gdp_pop_cleaned, x='Population', y='GDP', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Relación entre Población Total y Producto Interno Bruto (PIB) por País', fontsize=16)
plt.xlabel('Población Total', fontsize=12)
plt.ylabel('Producto Interno Bruto (PIB)', fontsize=12)
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

## <font color='green'>Conclusiones:</font></h1></center>

El gráfico de dispersión con ejes logarítmicos muestra una clara correlación positiva entre la Población y el PIB: los países con mayor población tienden a tener un PIB más alto. Sin embargo, la dispersión de los puntos también indica que la relación no es perfectamente lineal y que existen variaciones significativas en el PIB para países con poblaciones similares, lo que sugiere la influencia de otros factores económicos y estructurales más allá del simple tamaño de la población. Los países se agrupan principalmente en la parte inferior izquierda del gráfico, con algunos valores atípicos hacia la parte superior derecha (países con poblaciones y PIB muy elevados).

Resumen:

- Se usó escala logarítmica debido a la gran variación en los valores.

- Se eliminaron outliers extremos que podrían distorsionar la visualización.

- Se observa una relación positiva pero no perfecta entre población y PIB.

Analizaremos la relación entre **Tasa de Fertilidad** y **Tasa de Mortalidad Infantil:**

```python
df_fert_mort = df_filtrado[['Country', 'Fertility Rate', 'Infant mortality']].dropna()

df_fert_mort['Fertility Rate'] = pd.to_numeric(df_fert_mort['Fertility Rate'], errors='coerce')
df_fert_mort['Infant mortality'] = pd.to_numeric(df_fert_mort['Infant mortality'], errors='coerce')
df_fert_mort = df_fert_mort.dropna()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_fert_mort, x='Fertility Rate', y='Infant mortality')
plt.title('Tasa de Fertilidad vs Mortalidad Infantil')
plt.xlabel('Tasa de Fertilidad (hijos por mujer)')
plt.ylabel('Mortalidad Infantil (por 1000 nacimientos)')
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(data=df_fert_mort, x='Fertility Rate', y='Infant mortality')
plt.title('Relación entre Tasa de Fertilidad y Mortalidad Infantil')
plt.xlabel('Tasa de Fertilidad')
plt.ylabel('Mortalidad Infantil')
plt.show()

correlation = df_fert_mort['Fertility Rate'].corr(df_fert_mort['Infant mortality'])
print(f"Correlación entre Tasa de Fertilidad y Mortalidad Infantil: {correlation:.2f}")
```

## <font color='green'>Conclusiones:</font></h1></center>

- Existe una fuerte correlación positiva entre la tasa de fertilidad y la mortalidad infantil.

- Países con mayor tasa de fertilidad tienden a tener mayor mortalidad infantil, lo que podría indicar diferencias en desarrollo socioeconómico y acceso a salud.
